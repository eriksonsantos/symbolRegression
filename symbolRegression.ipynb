```python
import numpy as np
from gplearn.genetic import SymbolicRegressor
import matplotlib.pyplot as plt

def generate_data():
    x_data = np.random.uniform(-10, 10, size=50).reshape(-1, 1)
    y_data = 5*np.sin(x_data + 5) + x_data + 3
    return x_data, y_data

    return x_data, y_data

x_data, y_data = generate_data()

function_set = ['add', 'sub', 'mul', 'div',
                'sqrt', 'log', 'abs', 'neg', 'inv',
                 'sin', 'cos', 'tan']

est_gp = SymbolicRegressor(population_size=500, generations=20, stopping_criteria=0.01,
                           function_set=function_set,
                           p_crossover=0.7, p_subtree_mutation=0.1, p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1)

est_gp.fit(x_data, y_data.ravel())

print("Equação encontrada:")
print(est_gp._program)
```

```python
plt.scatter(x_data.flatten(),y_data)
```

```python
x_test = np.linspace(-30, 30, 100).reshape(-1, 1)
y_pred = est_gp.predict(x_test)

plt.scatter(x_data,y_data, label='Dados reais')
plt.plot(x_test, y_pred, color='red', label='Regressão simbólica')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Simbólica com Algoritmo Genético')
plt.show()
```

```python
import numpy as np
from gplearn.genetic import SymbolicRegressor
import matplotlib.pyplot as plt

def generate_data():
    x_data = np.random.uniform(-10, 10, size=50).reshape(-1, 1)
    y_data = 3*x_data**3 + 2*x_data**2 + 20*x_data + 3
    return x_data, y_data

    return x_data, y_data

x_data, y_data = generate_data()

function_set = ['add', 'sub', 'mul', 'div',
                'sqrt', 'log', 'abs', 'neg', 'inv',
                 'sin', 'cos', 'tan']

est_gp = SymbolicRegressor(population_size=500, generations=100, stopping_criteria=0.01,
                           function_set=function_set,
                           p_crossover=0.7, p_subtree_mutation=0.1, p_hoist_mutation=0.1, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1)

est_gp.fit(x_data, y_data.ravel())

print("Equação encontrada:")
print(est_gp._program)
```

```python
plt.scatter(x_data.flatten(),y_data)
```

```python
x_test = np.linspace(-20, 20, 100).reshape(-1, 1)
y_pred = est_gp.predict(x_test)

plt.scatter(x_data,y_data, label='Dados reais')
plt.plot(x_test, y_pred, color='red', label='Regressão simbólica')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Simbólica com Algoritmo Genético')
plt.show()
```
